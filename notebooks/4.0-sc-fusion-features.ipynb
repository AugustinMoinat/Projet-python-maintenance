# Fusion des caractéristiques

```python
import pandas as pd
```

```python
err = pd.read_csv("../data/prepared_data/erreurs.csv")
err
```

```python
maint = pd.read_csv("../data/prepared_data/maintenance.csv")
maint
```

```python
tele = pd.read_csv("../data/prepared_data/telemetrie.csv")
tele
```

```python
mach = pd.read_csv("../data/raw/PdM_machines.csv")
mach
```

```python
features = err.merge(maint,left_on=["machineID","datetime"],right_on=["machineID","datetime"]).merge(tele,left_on=["machineID","datetime"],right_on=["machineID","datetime"])
```

```python
print(features.columns)
features
```

On rajoute les caractéristiques des machines pour chaque ligne.

```python
features = features.merge(mach,on="machineID")
features
```

```python
features.to_csv("../data/prepared_data/features.csv",index=False)
```
