```python
import pandas as pd

error = pd.read_csv("PdM_errors.csv")
machines = pd.read_csv("PdM_machines.csv")
maintenance = pd.read_csv("PdM_maint.csv")
failures = pd.read_csv("PdM_failures.csv")
```

```python
telemetry = pd.read_csv("PdM_telemetry.csv")
```

Compter les erreurs, failures et maintenances de chaque machine

```python
import numpy as np
nb_error_par_machine = np.bincount(error["machineID"].values)[1:]
nb_failure_par_machine = np.bincount(failures["machineID"].values)[1:]
nb_maint_par_machine = np.bincount(maintenance["machineID"].values)[1:]
```

Créér les couleurs

```python
def color_map(model):
    if model == "model1":
        return "blue"
    elif model == "model2":
        return "green"
    elif model == "model3":
        return "red"
    else:
        return "orange"

colors = machines["model"].map(color_map)        
```

```python
import matplotlib.pyplot as plt

plt.scatter(machines["age"],nb_error_par_machine,c=colors)
plt.subplot(221) #la specification de la sous courbe doit être faite en spécifiant 3 chiffres
plt.scatter(machines["age"],nb_error_par_machine,c=colors)
plt.title("Age vs errors")
plt.subplot(222) 
plt.scatter(machines["age"],nb_failure_par_machine,c=colors)
plt.title("Age vs failure")
plt.subplot(223) 
plt.scatter(machines["age"],nb_maint_par_machine,c=colors)
plt.title("Age vs Maintenance")
plt.show()
```
